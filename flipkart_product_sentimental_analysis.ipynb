```python
import numpy as np

def read_matrix(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    # Extract matrix data from the file
    matrix_data = [list(map(float, line.strip().split(',')[1:])) for line in lines[1:]]
    matrix = np.array(matrix_data)
    return matrix

def write_matrix(file_path, matrix):
    with open(file_path, 'w') as file:
        # Write the first row with column labels
        file.write("#," + ",".join(map(str, range(matrix.shape[0]))) + "\n")
        
        # Write each row with data
        for i, row in enumerate(matrix):
            file.write(f"{i},{','.join(map(str, row))}\n")

def main():
    input_file_path = "inp.txt"
    output_file_path = "inv.txt"

    # Read the matrix from inp.txt
    original_matrix = read_matrix(input_file_path)

    # Calculate the inverse of the matrix using NumPy
    inverse_matrix = np.linalg.inv(original_matrix)

    # Round the inverse matrix to 3 decimal places
    rounded_inverse_matrix = np.round(inverse_matrix, 3)

    # Write the rounded inverse matrix to inv.txt
    write_matrix(output_file_path, rounded_inverse_matrix)

if __name__ == "__main__":
    main()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Vectorize the comments
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(df['Comment'])

# Apply LDA
lda = LatentDirichletAllocation(n_components=5, random_state=0)
lda.fit(X)

# Display topics
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

display_topics(lda, vectorizer.get_feature_names_out(), 10)
```

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

def filter_comments(df, rating=None, keyword='', sentiment=None):
    """
    Filter comments based on rating, keyword, and sentiment.
    
    Parameters:
    - df: DataFrame containing the data.
    - rating: Rating to filter comments (default None to include all ratings).
    - keyword: Word to search for in comments.
    - sentiment: 'positive' or 'negative' to filter sentiment (default None to include both).
    
    Returns:
    - Filtered comments that match the criteria as a list of tuples (rating, sentiment, comment).
    """
    # Filter by rating if provided
    if rating is not None:
        filtered_df = df[df['Rating'].astype(int) == rating]
    else:
        filtered_df = df
    
    # Filter by keyword
    keyword_filtered_df = filtered_df[filtered_df['Comment'].str.contains(keyword, case=False, na=False)]
    
    # Analyze sentiment if provided
    if sentiment is not None:
        if sentiment == 'positive':
            keyword_filtered_df['Sentiment'] = keyword_filtered_df['Comment'].apply(lambda x: analyzer.polarity_scores(x)['compound'] > 0)
        elif sentiment == 'negative':
            keyword_filtered_df['Sentiment'] = keyword_filtered_df['Comment'].apply(lambda x: analyzer.polarity_scores(x)['compound'] < 0)
        else:
            raise ValueError("Sentiment must be 'positive' or 'negative'")
        
        # Final filter by sentiment
        final_filtered_df = keyword_filtered_df[keyword_filtered_df['Sentiment']]
    else:
        # Analyze both sentiments if not provided
        keyword_filtered_df['Sentiment'] = keyword_filtered_df['Comment'].apply(
            lambda x: 'positive' if analyzer.polarity_scores(x)['compound'] > 0 else 'negative')
        
        # Return all comments with their sentiment labels
        final_filtered_df = keyword_filtered_df
    
    # Convert to list of tuples (rating, sentiment, comment)
    return final_filtered_df[['Rating', 'Sentiment', 'Comment']].values.tolist()

def print_filtered_comments(comments):
    """
    Print each comment line by line with rating and sentiment.
    
    Parameters:
    - comments: List of tuples containing (rating, sentiment, comment) to print.
    """
    if comments:
        print("Filtered Comments:")
        for rating, sentiment, comment in comments:
            print(f"Rating: {rating}, Sentiment: {sentiment}")
            print(comment)
            print()  # Print an empty line between comments
    else:
        print("No comments found matching the criteria.")

# Example usage
# Sample DataFrame
# data = {
#     'Customer Name': ['Alice', 'Bob', 'Charlie'],
#     'Review Title': ['Great printer', 'Not bad', 'Could be better'],
#     'Rating': [5, 3, 2],
#     'Comment': ['This printer is amazing and has great features.', 'The printer is okay but has some issues.', 'I am not satisfied with the printer quality.']
# }

# df = pd.DataFrame(data)

# Filter and print comments
comments = filter_comments(df, rating=None, keyword='', sentiment='negative')
print_filtered_comments(comments)
```
